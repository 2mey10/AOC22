# read data

```python
data = []
with open("input.txt") as f:
    data = f.readlines()
```

```python
data
```

```python
data_trimmed = [el.split("\n")[0] for el in data]
```

```python
data_trimmed
```

```python
data_inted = []
for el in data_trimmed:
    try:
        data_inted.append(int(el))
    except:
        data_inted.append(-1)
```

```python
data_inted
```

```python
best = 0
count = 0
for el in data_inted:
    if el == -1:
        count = 0
    else:
        count+=el

        if count > best:
            best = count

print(best)
```
